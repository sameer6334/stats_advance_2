### Hypothesis Testing in Statistics

# What is hypothesis testing in statistics?
"""
Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample data to infer that a certain condition is true for the entire population.
"""

# What is the null hypothesis, and how does it differ from the alternative hypothesis?
"""
The null hypothesis (H0) is a statement that assumes no effect or no difference exists. The alternative hypothesis (H1) is the opposite, stating that there is a significant effect or difference.
"""

# What is the significance level in hypothesis testing, and why is it important?
"""
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01. It helps control Type I errors.
"""

# What does a P-value represent in hypothesis testing?
"""
The P-value is the probability of obtaining results at least as extreme as the observed results, assuming the null hypothesis is true. It is used to determine statistical significance.
"""

# How do you interpret the P-value in hypothesis testing?
"""
- If P-value < α, reject the null hypothesis (statistically significant result).
- If P-value ≥ α, fail to reject the null hypothesis (insufficient evidence).
"""

# What are Type 1 and Type 2 errors in hypothesis testing?
"""
- Type 1 Error: Rejecting a true null hypothesis (false positive).
- Type 2 Error: Failing to reject a false null hypothesis (false negative).
"""

# What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
"""
- One-tailed test: Tests for an effect in only one direction.
- Two-tailed test: Tests for an effect in both directions.
"""

# What is the Z-test, and when is it used in hypothesis testing?
"""
A Z-test is used when the sample size is large (n > 30) and the population variance is known. It determines if there is a significant difference between sample and population means.
"""

# How do you calculate the Z-score, and what does it represent in hypothesis testing?
"""
Z-score = (X - μ) / σ
Where:
- X = sample mean
- μ = population mean
- σ = population standard deviation

It represents how many standard deviations a value is from the mean.
"""

# What is the T-distribution, and when should it be used instead of the normal distribution?
"""
The T-distribution is used instead of the normal distribution when the sample size is small (n < 30) and the population standard deviation is unknown.
"""

# What is the difference between a Z-test and a T-test?
"""
- Z-test: Used when population variance is known and sample size is large.
- T-test: Used when population variance is unknown and sample size is small.
"""

# What is the T-test, and how is it used in hypothesis testing?
"""
A T-test compares the means of two groups to determine if they are significantly different from each other.
"""

# What is the relationship between Z-test and T-test in hypothesis testing?
"""
Both tests compare sample means, but the T-test accounts for small sample sizes and unknown variances, while the Z-test assumes known population variance and large samples.
"""

# What is a confidence interval, and how is it used to interpret statistical results?
"""
A confidence interval provides a range of values within which the true population parameter is likely to fall, based on the sample data.
"""

# What is the margin of error, and how does it affect the confidence interval?
"""
The margin of error represents the range of uncertainty around the sample estimate. A smaller margin means more precise results.
"""

# How is Bayes' Theorem used in statistics, and what is its significance?
"""
Bayes' Theorem calculates conditional probabilities, helping update beliefs based on new evidence. It is widely used in machine learning and decision-making.
"""

# What is the Chi-square distribution, and when is it used?
"""
The Chi-square distribution is used in tests of independence and goodness-of-fit to determine if categorical data follows an expected distribution.
"""

# What is the Chi-square goodness of fit test, and how is it applied?
"""
This test determines if a sample distribution matches an expected distribution by comparing observed and expected frequencies.
"""

# What is the F-distribution, and when is it used in hypothesis testing?
"""
The F-distribution is used in variance analysis, such as ANOVA and the F-test, to compare multiple population variances.
"""

# What is an ANOVA test, and what are its assumptions?
"""
ANOVA (Analysis of Variance) tests whether there are significant differences between means of three or more groups. Assumptions include normality, independence, and equal variance.
"""

# What are the different types of ANOVA tests?
"""
- One-way ANOVA: Compares means across one independent variable.
- Two-way ANOVA: Compares means across two independent variables.
- Repeated Measures ANOVA: Used for dependent samples.
"""

# What is the F-test, and how does it relate to hypothesis testing?
"""
The F-test compares variances of two or more groups to determine if they are significantly different. It is commonly used in ANOVA.
"""

```python
### Hypothesis Testing in Statistics

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform a Z-test
def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    
    print(f"Z-Score: {z_score}")
    print(f"P-Value: {p_value}")
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
    
    return z_score, p_value

# Function to simulate random data and perform a Z-test
def simulate_z_test():
    np.random.seed(42)
    sample_data = np.random.normal(100, 15, 30)  # Mean=100, Std=15, n=30
    population_mean = 105
    population_std = 15
    
    return z_test(sample_data, population_mean, population_std)

# Function to visualize Type 1 and Type 2 errors
def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Normal Distribution')
    plt.fill_between(x, y, where=(x > 1.96), color='red', alpha=0.5, label='Type 1 Error')
    plt.fill_between(x, y, where=(x < -1.96), color='red', alpha=0.5)
    plt.fill_between(x, y, where=(x > -1.96) & (x < 1.96), color='blue', alpha=0.3, label='Type 2 Error')
    plt.legend()
    plt.title("Type 1 and Type 2 Errors")
    plt.show()

# Function to perform an independent T-test
def independent_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    print(f"T-Statistic: {t_stat}")
    print(f"P-Value: {p_value}")
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
    return t_stat, p_value

# Function to perform a paired T-test
def paired_t_test(before, after, alpha=0.05):
    t_stat, p_value = stats.ttest_rel(before, after)
    print(f"T-Statistic: {t_stat}")
    print(f"P-Value: {p_value}")
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
    return t_stat, p_value

# Function to calculate a confidence interval
def confidence_interval(sample_data, confidence=0.95):
    mean = np.mean(sample_data)
    std_err = stats.sem(sample_data)
    ci = stats.t.interval(confidence, len(sample_data)-1, loc=mean, scale=std_err)
    print(f"{confidence*100}% Confidence Interval: {ci}")
    return ci

# Example Execution
if __name__ == "__main__":
    simulate_z_test()
    visualize_errors()
    
    # Sample Data
    np.random.seed(42)
    sample1 = np.random.normal(100, 10, 30)
    sample2 = np.random.normal(105, 10, 30)
    independent_t_test(sample1, sample2)
    
    before = np.random.normal(100, 10, 30)
    after = before + np.random.normal(2, 5, 30)
    paired_t_test(before, after)
    
    confidence_interval(sample1)
```

```python
### Hypothesis Testing in Statistics

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to perform a Z-test
def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    
    print(f"Z-Score: {z_score}")
    print(f"P-Value: {p_value}")
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
    
    return z_score, p_value

# Function to simulate random data and perform a Z-test
def simulate_z_test():
    np.random.seed(42)
    sample_data = np.random.normal(100, 15, 30)  # Mean=100, Std=15, n=30
    population_mean = 105
    population_std = 15
    
    return z_test(sample_data, population_mean, population_std)

# Function to visualize Type 1 and Type 2 errors
def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Normal Distribution')
    plt.fill_between(x, y, where=(x > 1.96), color='red', alpha=0.5, label='Type 1 Error')
    plt.fill_between(x, y, where=(x < -1.96), color='red', alpha=0.5)
    plt.fill_between(x, y, where=(x > -1.96) & (x < 1.96), color='blue', alpha=0.3, label='Type 2 Error')
    plt.legend()
    plt.title("Type 1 and Type 2 Errors")
    plt.show()

# Function to perform an independent T-test
def independent_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    print(f"T-Statistic: {t_stat}")
    print(f"P-Value: {p_value}")
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
    return t_stat, p_value

# Function to perform a paired T-test
def paired_t_test(before, after, alpha=0.05):
    t_stat, p_value = stats.ttest_rel(before, after)
    print(f"T-Statistic: {t_stat}")
    print(f"P-Value: {p_value}")
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
    return t_stat, p_value

# Function to calculate a confidence interval
def confidence_interval(sample_data, confidence=0.95):
    mean = np.mean(sample_data)
    std_err = stats.sem(sample_data)
    ci = stats.t.interval(confidence, len(sample_data)-1, loc=mean, scale=std_err)
    print(f"{confidence*100}% Confidence Interval: {ci}")
    return ci

# Function to perform a Chi-square test for independence
def chi_square_test(observed):
    chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)
    print(f"Chi-Square Statistic: {chi2_stat}")
    print(f"P-Value: {p_value}")
    if p_value < 0.05:
        print("Reject the null hypothesis: Variables are dependent")
    else:
        print("Fail to reject the null hypothesis: Variables are independent")
    return chi2_stat, p_value, expected

# Function to perform an ANOVA test
def anova_test(*groups):
    f_stat, p_value = stats.f_oneway(*groups)
    print(f"F-Statistic: {f_stat}")
    print(f"P-Value: {p_value}")
    if p_value < 0.05:
        print("Reject the null hypothesis: At least one group mean is different")
    else:
        print("Fail to reject the null hypothesis: Means are similar")
    return f_stat, p_value

# Function to visualize ANOVA results
def visualize_anova(*groups):
    plt.figure(figsize=(8, 5))
    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title("ANOVA Test Results")
    plt.show()

# Example Execution
if __name__ == "__main__":
    simulate_z_test()
    visualize_errors()
    
    # Sample Data
    np.random.seed(42)
    sample1 = np.random.normal(100, 10, 30)
    sample2 = np.random.normal(105, 10, 30)
    independent_t_test(sample1, sample2)
    
    before = np.random.normal(100, 10, 30)
    after = before + np.random.normal(2, 5, 30)
    paired_t_test(before, after)
    
    confidence_interval(sample1)
    
    observed_data = np.array([[20, 30], [25, 35]])
    chi_square_test(observed_data)
    
    group1 = np.random.normal(100, 10, 30)
    group2 = np.random.normal(105, 10, 30)
    group3 = np.random.normal(110, 10, 30)
    anova_test(group1, group2, group3)
    visualize_anova(group1, group2, group3)
```
